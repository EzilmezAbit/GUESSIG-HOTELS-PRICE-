```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd
driver = webdriver.Edge()  # Edge tarayıcısı için webdriver kullanın
file = open("data.csv", "w", encoding="utf-8")
file.close()
data_list = []

for i in range(2,35):

    URL="https://www.trivago.com.tr/tr/srl/kad%C4%B1k%C3%B6y-yak%C4%B1nlar%C4%B1nda-oteller?search=500-1973723;dr-20240115-20240116;pa-"+str(i)
    driver.get(URL)

    time.sleep(8)


    for i in range(1,35):

        name_path = f"/html/body/div[1]/div[1]/main/div[5]/section/div/div/ol/li[{i}]/div/article/div[2]/div[1]/section/h2/button/span" 
        puan_path = f"/html/body/div[1]/div[1]/main/div[5]/section/div/div/ol/li[{i}]/div/article/div[2]/div[1]/button[2]/span[1]/span/span/strong/span" 
        fiyat_path=f"/html/body/div[1]/div[1]/main/div[5]/section/div/div/ol/li[{i}]/div/article/div[2]/div[2]/div[1]/div/div[3]/div[1]/div/span/p"
        uzaklık_path=f"/html/body/div[1]/div[1]/main/div[5]/section/div/div/ol/li[{i}]/div/article/div[2]/div[1]/button[1]/span[1]"
        try:       
            name_name = driver.find_element(By.XPATH, name_path).text
            puan_name = driver.find_element(By.XPATH, puan_path).text
            fiyat_name = driver.find_element(By.XPATH, fiyat_path).text
            uzaklık_name = driver.find_element(By.XPATH, uzaklık_path).text
        
       
        
            driver.execute_script("window.scrollTo(0,document.body.scrollHeight-1500)")
            time.sleep(0.50)
        
            data_list.append([name_name,puan_name,fiyat_name,uzaklık_name])
            print(name_name,puan_name,fiyat_name,uzaklık_name)                    

        except Exception as e:
            print(f"{i}. veri alınırken hata oluştu: {e}")

df = pd.DataFrame(data_list,columns=["name","puan","fiyat","uzaklık"])
df.to_csv("data.csv",index=False,encoding="utf-8")

driver.quit()
```
